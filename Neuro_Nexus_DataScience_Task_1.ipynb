## Using the macine learning model to predict the movie rating


### 1.) Splitting the data into X and Y

```python
X = cleaned_movies_data.drop("Rating", axis=1)
X = cleaned_movies_data.drop("Name", axis=1)
Y = cleaned_movies_data["Rating"]
```

```python
X.head(10)
```

```python
Y.head(10)
```

### 2.) converting the data that can be used for prediction into numerical values

```python
# importing scikit-learn libraries for converting the data to numerical values

from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer

categorical_features = ["Year", "Duration", "Genre", "Votes", "Director", "Actor 1", "Actor 2", "Actor 3"]
one_hot = OneHotEncoder()
transformer = ColumnTransformer([("one_hot",
                                 one_hot,
                                 categorical_features)],
                               remainder="passthrough")
transformed_X = transformer.fit_transform(X)
```

### 3.) Fitting the data to the model
    
    - We are predicting a quantity that's why we use a regression model
    - Since the dataset we are working on has less than 100k samples and some features are important that's why I have used Ridge model from the `sklear.linear_model library`

```python
# importing the libraries
 
from sklearn.linear_model import Ridge
from sklearn.model_selection import train_test_split

# splitting the data into train and test

X_train, X_test, Y_train, Y_test = train_test_split(transformed_X, Y, test_size=0.2)

# Instantiate and fit the model (on the training set)

model = Ridge()
model.fit(X_train, Y_train)

# checking for the score (on the testing set)

model.score(X_test, Y_test)
```

### 4.) Checking the Mean Absolute Error Values

```python
# importing the sklearn libraries

from sklearn.metrics import mean_absolute_error

# checking the mean absolute error

y_preds = model.predict(X_test)
mae = mean_absolute_error(Y_test, y_preds)
mae
```
